# create_cluster_folders

```python
import pandas as pd
import os
```

```python
CATALOG_PATH = '/home/aubrey/Desktop/tinian_sticky_trap_images/catalog.csv'
CLUSTERS_PATH = '/home/aubrey/Desktop/tinian_sticky_trap_images/clusters'
```

```python
pd.set_option('max_colwidth', 0)
df = pd.read_csv(CATALOG_PATH)
df
```

```python
for i, r in df.iterrows():
    basename = os.path.basename(r.path)
    from_path = r.path
    from_path_parts = from_path.split('/')
    to_path = f'{CLUSTERS_PATH}/{r.size_class}/cluster{r.cluster}/{from_path_parts[-2]}-{basename}'
    print(f'Copying from {from_path} to {to_path}')
    dirname = os.path.dirname(to_path)
    if not os.path.exists(dirname):
        os.makedirs(dirname)
    os.system(f'cp {from_path} {to_path}')
```
